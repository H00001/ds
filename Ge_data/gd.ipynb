```python
import csv
def get_csv(csv_file):
    account_list = []
    with open(csv_file, 'r') as csvfile:
        csvreader = csv.reader(csvfile)
        for row in csvreader:
            username, password = row
        # å°†è´¦å·å’Œå¯†ç æ”¾å…¥åˆ—è¡¨ä¸­
            account_list.append(
                {"username":username, 
                "password":password})
    return account_list

def get_data_from_file(filename):
    # æ‰“å¼€æ–‡ä»¶å¹¶è¯»å–å†…å®¹
    with open(filename, 'r') as f:
        lines = f.readlines()
    groups = [lines[i:i+3] for i in range(0, len(lines), 3)]
    result = []
    for group in groups:
        try:
            id_, name = group[0].strip(), group[1].strip()
            phone, gender, age, time1, time2 = group[2].strip().split('\t')
        except: 
            pass
        result.append({
            'id': id_,
            'name': name,
            'code': phone,
            'gender': gender,
            'age': age,
            'time1': time1,
            'time2': time2
        })
    if len(result) >= len(lines)/3:
        return result
    else:
        print("Error read length:%d, real:%d" % (len(result),len(lines)/3))
        print(result[len(result)-1])
dt = get_data_from_file("long.txt")
```

```python
from email.utils import formataddr
import smtplib
from email.mime.text import MIMEText
from email.header import Header

def send_mail(sender,pwd,receivers,title,file):
    f= open(file,encoding='utf-8')
    content=f.read()
    f.close()
    message = MIMEText(content,'html','utf-8')
    message['From'] = sender
    message['To'] =  receivers[0]
    message['Subject'] = Header(title, 'utf-8')
    
    try:
        smtpObj = smtplib.SMTP('smtp.163.com')
        smtpObj.login(sender,pwd)
        smtpObj.sendmail(sender, receivers, message.as_string())
        print("é‚®ä»¶å‘é€æˆåŠŸ")
        print(receivers)
    except smtplib.SMTPException:
        print("Error: æ— æ³•å‘é€é‚®ä»¶")

```
